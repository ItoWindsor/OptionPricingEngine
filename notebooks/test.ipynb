```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
S0 = 100
r = 0.02
q = 0.01
sigma = 0.2 
T = 1 

n = 1000
m = 2
```

```python
def generate_paths_black_scholes(S0: float, r : float,q: float, sigma: float, T:float, n_time : int, m_mc : int):
    t = np.linspace(0, T, n+1)
    h = T/n_time
    W = np.insert(np.random.normal(scale= np.sqrt(h), size=(m_mc, n_time)).cumsum(axis=1), 0, 0, axis=1)
    S = S0*np.exp((r-q - (sigma**2)/2)*t + sigma*W)
    return S
```

```python
S = generate_paths_black_scholes(S0, r, q, sigma, T, n, m)
```

```python
t = np.linspace(0,T,n+1)
for k in range(m):
    plt.plot(t,S[k,:])
    
plt.grid()
plt.xlabel("time");plt.ylabel("stock price")
plt.show()
```
